Data Overview:

Telco Customer Churn dataset is sourced from Kaggle. The raw data contains around 7043 records and 21 variables. The predictive variable here is Churn where 1 represents those customers who did churn and 0 represents the inverse case. The remaining independent variables provide context around expenses of the subscribed plans, demographic information of the customers, length of current subscription, type of plan, payment type information, and additional information about the benefits of the subscription.

The data description of those variables is provided below:

●	Customer ID
    -The unique identifier within the dataset, this is the primary key
    
●	Gender
	-Categorical classifier representing Male/Female
    
●	Senior Citizen
	-Is the subscriber 65 years old or above?
    
●	Partner
	-Whether the subscriber has a life partner
    
●	Dependents
	-Whether the subscriber supports others financially
    
●	Tenure
	-Length of continuous subscription in months
    
●	Phone Service
	-Whether the plan includes phone services
    
●	Multiple Lines
	-Whether the plan has multiple phone line services
    
●	Internet Service
	-Whether the plan includes Internet Service
    
●	Online Security
	-Whether the plan includes this service
    
●	Online Backup
	-Whether the plan includes this service
    
●	Device Protection
	-Whether the plan includes this service
    
●	Tech Support
	-Whether the plan includes this service
    
●	Streaming TV
	-Whether the plan includes this service
    
●	Streaming Movies
	-Whether the plan includes this service
    
●	Contract
	-Type of fixed-term contract or if a month-to-month subscription
    
●	Paperless Billing
	-Whether the plan includes this service
    
●	Payment Method
	-How the subscriber pays their bill
    
●	Monthly Charges
	-Amount subscriber pays monthly
    
●	Total Charges
	-Sum of charges over the life of continuous service

#### d. Tenure, monthly charges and total charges

```python
# Tenure is recorded in months. To help gain better insights, this attribute is binned.
df['tenure'].describe()
```

```python
# As observed from above, tenure goes from 0 to 72 months. Hence binning them into five groups as below

def decode(col) :
    
    if col <= 12 :
        return "Tenure_0-12"
    elif (col > 12) & (col <= 24 ):
        return "Tenure_12-24"
    elif (col > 24) & (col <= 48) :
        return "Tenure_24-48"
    elif (col > 48) & (col <= 60) :
        return "Tenure_48-60"
    elif col > 60 :
        return "Tenure_gt_60"
    
df['tenure_group'] = df['tenure'].apply(decode)
df['tenure_group'].value_counts()
```

```python
# distribution of monthly charges for each tenure group 

sns.boxplot(x='tenure_group',y='MonthlyCharges',data=df,hue='Churn')
plt.tight_layout()
```

```python
# Clearly for every group, customers who churned were paying more monthly charges although there are a few outliers.
```

```python
# distribution of total charges for each tenure group

sns.boxplot(x='tenure_group',y='TotalCharges',data=df,hue='Churn')
plt.tight_layout()
```

## Data preparation for Modeling

#### 1. Replacing totalcharges of value zero to the avg totalcharges based on tenure_group and contract type

```python
df[df['TotalCharges']==0][['Contract','tenure_group']]
```

```python
df[(df['Contract']=='Two year') & (df['tenure_group']=='Tenure_0-12')]['TotalCharges'].mean()
```

```python
df[(df['Contract']=='One year') & (df['tenure_group']=='Tenure_0-12')]['TotalCharges'].mean()
```

```python
def impute(cols):
    totalcharges = cols[0]
    contract = cols[1]
    if (totalcharges==0):
        if (contract == 'Two year'):
             return 185.81
        else:
             return 300.723
    else:
        return totalcharges
df['TotalCharges'] = df[['TotalCharges','Contract']].apply(impute,axis=1)
```

```python
df.describe()
```

#### 2. Decoding Churn to 1 for Yes and 0 for No

```python
df['Churn'] = df['Churn'].replace({'Yes':1,'No':0})
```

#### 3. Creating dummies for all other categorical features

```python
final_df = df.drop(['customerID'],axis=1)
final_df = pd.get_dummies(data = final_df)
```

```python
final_df.shape
```

```python
final_df.columns
```

## Feature Engineering

Creating a new feature called no_of_services which is to count the number of services taken by customer

```python
fn = lambda row:row.PhoneService_Yes + row.MultipleLines_Yes + row.OnlineSecurity_Yes + row.OnlineBackup_Yes + row.DeviceProtection_Yes + row.TechSupport_Yes+ row.StreamingTV_Yes+ row.StreamingMovies_Yes
final_df['no_of_services'] = final_df.apply(fn,axis=1)
```

```python
final_df.head()
```

```python
final_df['no_of_services']=final_df['no_of_services'].astype(int)
```

```python
sns.countplot(x='no_of_services',data=final_df,hue='Churn',palette='coolwarm')
```
